#### Automatic Version, but you need to adjust a little bit when it's in your case

```python
from Ashare import *
import pandas as pd
import os

# Function to save checkpoint
def save_checkpoint(index):
    with open('checkpoint.txt', 'w') as checkpoint_file:
        checkpoint_file.write(str(index))

# Function to load checkpoint
def load_checkpoint():
    if os.path.exists('checkpoint.txt'):
        with open('checkpoint.txt', 'r') as checkpoint_file:
            return int(checkpoint_file.read().strip())
    else:
        return 0

# Open the text file containing stock codes and read each line
stock_data = []
start_flag = True
# Load the checkpoint (last successfully processed stock code)
last_processed_stock_code = load_checkpoint()

with open('stock_code.txt', 'r', encoding='utf-8') as file:
    # Iterate through the file starting from the last successfully processed stock code
    for i, line in enumerate(file):
        # Remove newline character from the end of each line
        stock_code = line.strip()

        # Check if we should start fetching data
        if start_flag:
            print("Fetching stock data for:", stock_code)

            try:
                # Fetch stock data using the get_price function
                df = get_price(stock_code, frequency='1d', count=50000, end_date='2024-04-30')

                # Inform about success or failure of data fetching
                if df is not None:
                    print("Successfully fetched stock data for:", stock_code)
                    # Append fetched data to the list
                    stock_data.append(df)
                    # Save the checkpoint after successfully processing each stock code
                    save_checkpoint(stock_code)
                else:
                    print("Failed to fetch stock data for:", stock_code)
            
            except Exception as e:
                print("An error occurred while fetching stock data:", e)
                print("Current stock code:", stock_code)
                break  # Exit the loop if an error occurs

        # Check if the current line matches the last successfully processed stock code
        if stock_code == last_processed_stock_code:
            start_flag = True
```
